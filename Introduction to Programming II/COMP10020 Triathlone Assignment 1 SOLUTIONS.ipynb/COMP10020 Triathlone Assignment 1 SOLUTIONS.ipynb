Often referred to as **sabremetrics**, analysis of sports data has become a huge application domain for data science. Anaysing sports data allows teams organise themselves better, athletes better understand what drives their performance, and fans get closer to the game. In this assignment we will anayse data from the 2017 Triathlone Sprint Triathlon event [Triathlone](http://www.triathlone.com/) to understand what drives triathlon success. A triathlon is a multi-sport event in which atheletes complete a **swim**, a **cycle** and a **run** in the fastest time possible. Between the swim and cycle and cycle and run there are timed **transitions** during which competitors change equipment.

The following code block reads data from a data file (*'TriathloneData.csv'*) scraped from the Triathlone website and stores this in a list of dictionary objects, each of which contains the details of a competitor's race result. The details stored about each result are stored in a dictionary object with the following keys:

* **Place:** The place in which the athlete finished the race (missing for non-finishers)
* **Number:** The athlete's race bib number
* **Wave:** The wave with which the athlete started (one of 1, 2, or 3)
* **Age_Cat:** The athlete's age category (one of 16-19, 20-29, 30-39, 40-49, or 50+)
* **Gender:** The gender that the athlete declared (one of 'M' or 'F')
* **TI_Number:** Some athlete's are members of the Triathlon Ireland association and if so declare their membership number	
* **Swim:** The time taken for the swimming leg of the event (in format hh:mm:ss)
* **T1:** The time taken for the first transition of the event (in format hh:mm:ss)
* **Cycle:** The time taken for the cycling leg of the event (in format hh:mm:ss)
* **T2:** The time taken for the swimming leg of the event (in format hh:mm:ss)
* **Run:** The time taken for the running leg of the event (in format hh:mm:ss)
* **Finish:** The time taken for the total event (in format hh:mm:ss)

```python
# Load TriathloneData.csv
results = list()
count = 0
with open('TriathloneDataNoHeader.csv') as f:
    for line in f:
        line = line.strip()
        words = line.split(',')
        result = {"Place":int(words[0]),"Number":words[1],"Wave":words[2],"Age_Cat":words[3],"Gender":words[4],"TI_Number":words[5],"Swim":words[6],"T1":words[7],"Cycle":words[8],"T2":words[9],"Run":words[10],"Finish":words[11]}
        results.append(result)
        count = count + 1
    
print(str(count) + " results loaded")
```

## Question 5: 

The times in the datset are written as strings in the format "hh:mm:ss". For example, the swiim time of "00:08:38" is 8 minutes and 38 seconds. Write a function to convert the string times into seconds. For example, the swim time of "00:08:38" should convert to 518 seconds.

```python
import numpy as np
def hms_to_seconds(t):
    if(type(t) != str):
        return np.NaN
    else:
        h, m, s = [int(i) for i in t.split(':')]
        return 3600*h + 60*m + s
```

The following piece of code will add new elements to each result dictioanry storing the race times in seconds.

```python
for result in results:
    result['Swim_secs'] = hms_to_seconds(result['Swim'])
    result['T1_secs'] = hms_to_seconds(result['T1'])    
    result['Cycle_secs'] = hms_to_seconds(result['Cycle'])
    result['T2_secs'] = hms_to_seconds(result['T2'])    
    result['Run_secs'] = hms_to_seconds(result['Run'])
    result['Finish_secs'] = hms_to_seconds(result['Finish'])
    
results
```

## Question 6:

There was a problem with the timing system used on the day of the race. All participants were mistakenly given their swim time as their first transition (**T1**) time. All other times arre correct. Calculate a new corrected T1 time for each particcpant and add it to the dictionary.

```python
for result in results:
    total_time_without_T1 = result['Swim_secs']  + result['Cycle_secs'] + result['T2_secs'] + result['Run_secs']
    T1_time = result['Finish_secs'] - total_time_without_T1
    result['T1_corrected'] = T1_time
results
```
