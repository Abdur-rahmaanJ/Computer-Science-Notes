## A simple interaction

```python
# Rough graphics..
r = int(input("Number of rows: "))
c = int(input("Number of columns: "))
matrix = [[0 for x in range(c)] for y in range(r)]

for i in range(r):
    for j in range(c):
        matrix[i][j] = i*j

for i in range(r):
    for j in range(c):
        if matrix[i][j]<50:
            print ".",
        else:
            print "X",
    print("\n")
```
